```python
#Importing necessary libraries
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

Data collection and data pre-processing

```python
#Loading the dataset to the pandas dataframe
sonardata=pd.read_csv('/content/Copy of sonar data.csv',header=None)   # As there is no header in our dataset , that is why we mentioned header=None
```

```python
sonardata.head()    # We have R for Rock data and M for mine data
```

```python
#no of rows and columns
sonardata.shape
```

```python
#Getting statistical definition of data
sonardata.describe()
```

```python
sonardata[60].value_counts()  #We have 111 instances of mines and 97 instances for Rock
```

```python
sonardata.groupby(60).mean()  #We have found mean value of each feature for rock data and mine data.
```

```python
#Separating data and labels
X=sonardata.drop(columns=60,axis=1)   #X is a data
Y=sonardata[60]                       #Y is a label
```

```python
print(X)    #Printing data
```

```python
print(Y)    #Printing Label
```

Spliting data into training and test data.

x_train is training data.

x_test is testing data.

y_train is labels of training data.

y_test is labels of testing data.

test_size=0.15 means 15% data is used for testing the model.

stratify=Y means to split the data in train and test equally without bias.

random_state=1 means it allows you to reproduce the same train_test_split everytime you run the code.

```python
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=15,stratify=Y,random_state=1)
```

```python
print(x_train)
```

```python
print(y_train)
```

```python
print(X.shape)    #It has 208 rows with 60 columns
print(x_train.shape)  #Train data has 156 rows and 50 columns
print(x_test.shape)   #Test data has 52 rows and 60 columns
```

Defining logistic regression model and traing the model on the splitted dataset earlier.

```python
model=LogisticRegression()  #This will load the logistic regression model in variable model.
```

```python
#Training the logistic regressioon model on training data.
model.fit(x_train,y_train)
```

Model Evaluation of the model.

```python
#Accuracy on training data
x_train_prediction=model.predict(x_train)
trainingdataaccuracy=accuracy_score(x_train_prediction,y_train)   #Here we are comparing prediction of our model with the actual value i.e. y_train
```

```python
print('Accuracy Score on Training Data is: ',trainingdataaccuracy)
```

Here we have got accuracy of 83% on training data.

```python
#Accuracy on testing data
x_test_prediction=model.predict(x_test)
testingdataaccuracy=accuracy_score(x_test_prediction,y_test)    #Here we are comparing prediction of our model with test data i.e.y_test
```

```python
print('Accuracy Score on Testing Data is: ',testingdataaccuracy)
```

Here we have got accuracy of 73% on test data.
